```python
silver_catalog    = "silver_supplier"
gold_catalog = "gold_supplier"
gold_schema = "supplier_schema"
silver_schema     = "supplier_schema"
silver_supplier_basic_AI_csv = "silver_supplier_basic_AI_csv"
silver_supplier_feedback_AI_csv = "silver_supplier_feedback_AI_csv"
silver_supplier_feedback_AI_join_csv = "silver_supplier_feedback_AI_Join_csv"
gold_supplier_feedback_AI_join_csv = "gold_supplier_feedback_ai_join_csv"
```

```python
spark.sql("create catalog if not exists gold_supplier") 
spark.sql("create schema if not exists gold_supplier.supplier_schema") 
```

```python
%sql
drop table if exists silver_supplier.supplier_schema.silver_supplier_feedback_ai_join_csv;
drop table if exists gold_supplier.supplier_schema.gold_supplier_feedback_ai_join_csv
```

## Create the Join table if it does not exist ##

```python
spark.sql("create table if not exists silver_supplier.supplier_schema.silver_supplier_feedback_AI_Join_csv (supplier_name_silver  STRING, supplier_continent_silver STRING, Supplier_country_silver STRING, supplier_city_silver STRING, Summary_silver STRING, Rating STRING )")
spark.sql("create table if not exists gold_supplier.supplier_schema.gold_supplier_feedback_AI_Join_csv (supplier_name_gold  STRING, supplier_continent_gold STRING, Supplier_country_gold STRING, supplier_city_gold STRING, Summary_gold STRING, Rating STRING )")
```

## Create a new table which is the result of join between the table that has the "Continent" and the table that has "summary feedback" and "rating" ##

```python
%sql

insert into silver_supplier.supplier_schema.silver_supplier_feedback_ai_join_csv (supplier_name_silver  , supplier_continent_silver , Supplier_country_silver , supplier_city_silver , Summary_silver, Rating)
 SELECT
      t1.supplier_name_silver, t1.continent, t1.supplier_country_silver, t1.supplier_city_silver, t2.summary_silver, t2.rating
    FROM
       silver_supplier.supplier_schema.silver_supplier_basic_ai_csv  t1
    INNER JOIN
        silver_supplier.supplier_schema.silver_supplier_feedback_ai_csv t2 ON t1.supplier_name_silver = t2.supplier_name_silver;
```

```python
%sql
select * from silver_supplier.supplier_schema.silver_supplier_feedback_ai_join_csv
```

```python
%sql
insert into gold_supplier.supplier_schema.gold_supplier_feedback_ai_join_csv (supplier_name_gold  , supplier_continent_gold , Supplier_country_gold , supplier_city_gold , Summary_gold, Rating)
 SELECT
      t1.supplier_name_silver, t1.continent, t1.supplier_country_silver, t1.supplier_city_silver, t2.summary_silver, t2.rating
    FROM
       silver_supplier.supplier_schema.silver_supplier_basic_ai_csv  t1
    INNER JOIN
        silver_supplier.supplier_schema.silver_supplier_feedback_ai_csv t2 ON t1.supplier_name_silver = t2.supplier_name_silver;
```
