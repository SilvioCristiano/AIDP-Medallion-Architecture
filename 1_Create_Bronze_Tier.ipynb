


## **Read a file from the Supplier Catalog, rename some of the columns, and save it to bronze_supplier catalog.**

```python
# set up some variables 
catalog = "supplier"
schema = "supplier_schema"
table_basic = "basic_supplier"
table_emotions = "supplier_emotions"
bronze_catalog    = "bronze_supplier"
bronze_schema     = "supplier_schema"
bronze_table_csv = "bronze_supplier_basic_csv"
bronze_table_feedback_csv  = "bronze_supplier_feedback_csv"


```

## **Create the bronze catalog and schema if they do not exist** ##

```python


spark.sql("create catalog if not exists bronze_supplier") 
spark.sql("create schema if not exists bronze_supplier.supplier_schema") 
```

## Read the table file into a DataFrame ##

```python
bronze_df = spark.read.table(f"{catalog}.{schema}.{table_basic}")
bronze_df.show()
```

## Write the data frame into a bronze catalog table ## 

```python
 bronze_df.write.mode("overwrite").format("csv").saveAsTable(f"{bronze_catalog}.{bronze_schema}.{bronze_table_csv}")
   
```
